## Оценивание и штрафы



Групповой проект оценивается исходя из устной защиты (презентации) командой выполненного ими проекта перед комиссией. Комиссия составляется из преподавателей курса, ассистентов курса, а также приглашенных экспертов из индустрии.

В рамках каждой защиты комиссия выставляет два типа оценок.

Первая оценка — это оценка для всего проекта в целом. Она ставится комиссией, исходя из глубины проработки командой предметной области, сложности применяемых в проекте инструментов, а также прочих аспектов, связанных с содержанием выполняемого в проекте ТЗ. При выставлении этой оценки комиссия ориентируется на критерии, более подробно представленные ниже в разделе "Критерии оценки проекта".

Вторая оценка — это индивидуальная оценка каждого участника команды. Именно эта оценка выставляется впоследствии в ведомость. Индивидуальная оценка участника команды определяется комиссией, исходя из того, насколько участник ориентируется в выполненном командой проекте, как он отвечает на адресованные персонально ему вопросы комиссии, насколько в принципе прослеживается его роль в проекте, а также исходя из других аспектов, связанных с участием конкретного человека в деятельности команды. Индивидуальная оценка не может превосходить оценку команды за проект в целом (первую оценку), однако, в случае когда участник спокойно и корректно отвечает на все адресованные ему вопросы, свободно ориентируется в материале и содержании выполненного проекта, а также когда его роль в работе команды не вызывает сомнений, — в таком случае, индивидуальная оценка участника становится равной оценке для всего проекта (т.е. в идеальном случае — вторая оценка равна первой оценке).

Во избежание субъективности и/или предвзятости оценки тех или иных членов комиссии в отношении тех или иных команд, а также каких-либо иных проявлений неравных условий — распределение команд между комиссиями осуществляется абсолютно рандомно с использованием средств библиотеки Numpy :)) — в качестве рэндом сида выступает дата, в которую производится распределение, например, `np.random.seed(28012025)`.

Обратите, пожалуйста, внимание на то, что бОльшая часть комиссий осуществляет приём проектов в очной форме, однако могут присутствовать и 1-2 комиссии, которые будут работать дистанционно с использованием средств видеосвязи. В случае попадания команды путем случайного распределения в комиссию, работающую в очной форме, *не допускается* какое-либо изменение формата проведения для команды на онлайн-режим, равно как и любые другие перемещения команд между комиссиями (смены комиссий), для сохранения честности и непредвзятости случайного распределения.

2. Случайный лес
---

```python
#Название шоу будет мало ползным признаком для модели, а кодировать с помощью OHE не вариант, поэтому просто удалим.
data_rf = data.copy()
```

```python
data_rf
```

```python
#Выделим в качестве тренировочного датасета данные, у которых нет пропусков в таргете
train_df = data_rf.dropna(subset='user rating score')
y_train = train_df['user rating score']
X_train = train_df.drop(['user rating score', 'title'], axis=1)

#В качестве тестовых данных выделим строки с пропусками с столбце таргета
test_df = data_rf[data_rf['user rating score'].isna()]
X_test = test_df.drop(['user rating score', 'title'], axis=1)
y_test = test_df['user rating score']


```

```python
ct = ColumnTransformer(
    [
        ("ratingLevel_tfidf", TfidfVectorizer(), 'ratingLevel'),
        ("rating_ohe", OneHotEncoder(handle_unknown='ignore'), ['rating']),
        ("passthrough", "passthrough", ['release year']),
    ]
)

model = Pipeline([
    ("preprocessor", ct),
    ("rf", RandomForestRegressor(
    n_estimators = 50,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42
)),
])

#Теперь обучаем
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
#Теперь обратно склеим данные
test_df_filled = test_df.copy()
test_df_filled['user rating score'] = y_pred

#Соединяем тестовый и тренировочный датасет
data_no_miss = pd.concat([train_df, test_df_filled], ignore_index = True).sort_index()
```

```python
fig = px.histogram(
    data_no_miss,
    x="user rating score",
    nbins=80,
    title="Распределение оценок пользователей",
    labels={"user rating score": "Оценка пользователей", "count": "Количество оценок"},
    template="plotly_white",
)

fig.update_traces(
    marker=dict(
        line=dict(width=1, color="black",)))

fig.update_layout(
    xaxis_title="Оценка пользователей",
    yaxis_title="Количество оценок",
)

fig.show()
```

###Почему люди склонны ставить завышенные оценки?
По гистограмме распределения можно выдвинуть гипотезу, что люди склонны оценивать большинство шоу в диапазоне от 80 до 100. Это можно объяснить несколькими гипотезами:

1.   Предвзятость пользователей:
    *   Пользователи, которым понравилось шоу склонны скорее оставлять положительный отзыв, чтобы посоветовать это остальным людям, в свою очередь люди с негативным опытом просмотра скорее склонны вообще не оставлять никакого отзыва. Из этого следует, что представленная выборка оценок может быть предвзятой, так как людей, которые оставили положительные отзывы больше, чем людей, оставивших негативный отзыв.
2. Эмоциональный эффект:
    *  Люди, которые только посмотрели ТВ-шоу, на эмоциях от увиденного скорее склонны либо завышать оценки, как в нашем датасете, либо занижать.

```python
import time
from bs4 import BeautifulSoup
import requests
from wordcloud import WordCloud, STOPWORDS
import re
```

```python
#Данная функция будет искать жанр на Википедии
def get_genre(show_name):
    url = "https://en.wikipedia.org/wiki/" + show_name
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    capi = soup.find_all("td",class_="category")
    fill_val = [re.sub(r"\[\d+\]", "", i.get_text().strip().replace("\n",", ")) for i in capi]
    return fill_val
```

```python
#В данной же функции мы будем обращаться к сайту Rotten Tomatoes
def get_info_rt(url_val):
    url_mov ="m/"
    url_tv = "tv/"
    url = "https://www.rottentomatoes.com/"

    page = requests.get(url + url_mov + url_val)
    soup = BeautifulSoup(page.content, 'html.parser')

    capi = soup.select("ul.content-meta.info li")
    fill_val = [i.find(class_="meta-value").get_text().replace("\n","").strip().replace("and",",").replace(" ","") for i in capi]
    fill_lbl = [i.find(class_="meta-label").get_text().replace("\n","").strip(":").replace("and",",").replace(" ","") for i in capi]
    final_ser = pd.Series(fill_val,index=fill_lbl)

    if(len(final_ser) <= 1):
        page = requests.get(url + url_tv + url_val)
        soup = BeautifulSoup(page.content, 'html.parser')

        cont = soup.select("div.panel-body.content_body td")
        fill_lbl = [i.get_text().replace(" \n","").strip(": \n").replace(" ","") for i in cont][::2]
        fill_val = [i.get_text().replace(" \n","").strip(": \n").replace(" ","") for i in cont][1::2]
        final_ser = pd.Series(fill_val,index=fill_lbl)
        return final_ser

    return final_ser
```

```python
#С помощью данного скрипта будем создавать новый признак Genre
t1 = time.time()

data['Genre'] = "not found"
list_titles =  [str(i).strip().replace(" ","_") for i in data['title']]

change_titles = ['_(TV_series)','_(American_TV_series)','_(franchise)']

songs_to_avoid = "alternative rock ska pop funk hip hop electronic film score jazz classical orchestra country feature film soundtrack hindi"


for i in range(len(data.index[:])):
        wiki_status = 0
        fill_val = get_genre(list_titles[i])


        if len(fill_val) > 0:
            test_val = fill_val[0].lower().split(",",1)[0]

            x = test_val.strip() not in songs_to_avoid

            if x:
                wiki_status = 1
                data['Genre'][i] = " ".join(fill_val)
                continue

        for j in change_titles:
            changed_title = list_titles[i] + j
            fill_val2 = get_genre(changed_title)


            if len(fill_val2) > 0:
                wiki_status = 1
                data['Genre'][i] = " ".join(fill_val2)
                break

        if wiki_status == 0 :
            value_change = list_titles[i].lower()
            from_rt = get_info_rt(value_change)
            if len(from_rt)>0:
                data['Genre'][i] = from_rt.get(key = 'Genre')

            else:
                year = str(data['release year'][i])
                value_change2 = value_change + "_" + year
                from_rt2 = get_info_rt(value_change2)
                data['Genre'][i] = from_rt2.get(key = 'Genre')


t2= time.time()
time_ = t2-t1
print("Done in seconds : ",time_)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Отделяем Prison Break от остальных сериалов
prison_break = data[data["title"] == "Prison Break"]
other_shows = data[data["title"] != "Prison Break"]

fig = make_subplots(rows=2, cols=2, subplot_titles=[
    "Сравнение рейтинга Prison Break с другими сериалами",
    "Сравнение года выхода Prison Break с другими сериалами",
    "Популярные жанры и место Prison Break среди них",
    "Сравнение возрастных ограничений Prison Break с другими сериалами"])

#График гистограмма - сравнение по рейтингу с другими сериалами
hist_data = go.Histogram(x=other_shows["user rating score"], nbinsx=20, opacity=0.6, name="Другие сериалы")
prison_break_line = go.Scatter(x=[prison_break["user rating score"].values[0]]*2,
                               y=[0, 50], mode="lines", line=dict(color="red", dash="dash"), name="Prison Break")
fig.add_trace(hist_data, row=1, col=1)
fig.add_trace(prison_break_line, row=1, col=1)

#График сравнения по дате выхода
hist_data2 = go.Histogram(x=other_shows["release year"], nbinsx=20, opacity=0.6, name="Другие сериалы")
prison_break_line2 = go.Scatter(x=[prison_break["release year"].values[0]]*2,
                                y=[0, 50], mode="lines", line=dict(color="red", dash="dash"), name="Prison Break")
fig.add_trace(hist_data2, row=1, col=2)
fig.add_trace(prison_break_line2, row=1, col=2)

#К какому жанру оносится Prison break и насколько этот жанр популярен.
prison_break_genres = prison_break["merged_genre"].values[0].split(", ") if pd.notna(prison_break["merged_genre"].values[0]) else []
genre_counts = other_shows["merged_genre"].dropna().str.split(", ").explode().value_counts()

bar_data = go.Bar(x=genre_counts.index[:10], y=genre_counts.values[:10], name="Популярные жанры")
fig.add_trace(bar_data, row=2, col=1)

for genre in prison_break_genres:
    if genre in genre_counts.index[:10]:
        fig.add_trace(go.Bar(x=[genre], y=[genre_counts[genre]], marker_color='red', name="Prison Break"), row=2, col=1)

#Сравнение с другими сериалами по возрастной группе
age_group_counts = other_shows["Age_group"].value_counts()
bar_data2 = go.Bar(x=age_group_counts.index, y=age_group_counts.values, name="Другие сериалы")
fig.add_trace(bar_data2, row=2, col=2)

fig.add_trace(go.Bar(x=[prison_break["Age_group"].values[0]],
                      y=[age_group_counts.get(prison_break["Age_group"].values[0], 0)],
                      marker_color='red', name="Prison Break"), row=2, col=2)

fig.update_layout(title_text="Сравнительный анализ Prison Break", showlegend=False, template="plotly_white", width=1200, height=800)

fig.show()
```
